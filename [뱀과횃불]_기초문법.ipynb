# 복사와 결측치

```python
df['age'] = df['age'].fillna(30)
```

```python
assert df['age'].isnull().sum() == 0
assert df['age'].mean().round(4) == 29.7589
```

```python
df = sns.load_dataset('titanic')
```

```python
df1 = df.copy()
c1 = df1['sex'] == 'male'
c2 = df1['sex'] == 'female'
df1.loc[c1, 'age'] = df1.loc[c1, 'age'].fillna(df1.loc[c1, 'age'].mean())
df1.loc[c2, 'age'] = df1.loc[c2, 'age'].fillna(df1.loc[c2, 'age'].mean())
```

```python
assert (df1['age'].isnull().sum() == 0)
assert df1['age'].mean().round(5) == 29.73603
```

# Groupby와 Pivot Table

# 새 섹션

```python
sample.groupby('pclass')['survived'].mean()
```

```python
sample.groupby('embarked')['survived'].agg(['mean', 'var'])
```

```python
sample.groupby(['who', 'pclass'])['survived'].agg(['mean', 'sum'])
```

```python
male_c = sample['sex'] == 'male'
female_c = sample['sex'] == 'female'
male_age = sample.loc[sample['sex'] == 'male', 'age'].mean()
female_age = sample.loc[sample['sex'] == 'female', 'age'].mean()
sample.loc[male_c, 'age'] = sample.loc[male_c, 'age'].fillna(male_age)
sample.loc[female_c, 'age'] = sample.loc[female_c, 'age'].fillna(female_age)
```

```python
print(sample['age'].isnull().sum())
print(f"age 평균: {sample['age'].mean():.2f}")
```
